# Single Neuron Usng Numpy

image.png

````python
# Define two vectors
inputs = np.array([1, 2, 3])
weights = np.array([4, 5, 6])

bias = 2

# Dot product
result = np.dot(inputs, weights) + bias
# or
result = inputs @ weights + bias

print(result)
````

# Layer Of Neuron

image.png

````python
# Define a vector and a matrix
inputs = np.array([1, 2, 3, 4])
weights = np.array([[1, 2, 3, 4],
              [4, 5, 6, 9],
              [7, 8, 9, 20]])

## In the weights matrix , each row denotes the weights corresponding to a single neuron

bias = np.array([0.1,0.2,0.3])
bias = np.transpose(bias)

# Dot product
result = np.dot(weights, inputs) + bias
# or
result2 = np.dot(inputs, np.transpose(weights)) + np.transpose(bias)

print(result)  # Output: [30 36 42]
print(result2)
````
